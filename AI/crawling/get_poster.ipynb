```python
import pandas as pd
import os

print(os.getcwd())

df = pd.read_csv('db_dataset.csv', encoding='utf-8')
posters = df['images']
backdrops= df['backdrops']
```

```python
# from bs4 import BeautifulSoup
from urllib.request import urlopen
from urllib.parse import quote
# import requests

for i in range(18737,len(backdrops)):
    print(f'=============================== {i} ===================================')
    if backdrops[i] == '0':
        continue
    url = str(backdrops[i])
    url = url[:url.rfind('/')+1] + quote(url[url.rfind('/')+1:])
    print(url)
    # webpage = requests.get(url)
    # soup = BeautifulSoup(webpage.content, 'html.parser')
    try:
        with urlopen(url) as f:
            with open(f'../../../backdrops/{i}.jpg', 'wb') as h:
                img = f.read()
                h.write(img)
    except:
        continue    
```
