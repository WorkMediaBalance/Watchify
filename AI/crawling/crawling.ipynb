```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager


driver = webdriver.Chrome(executable_path=ChromeDriverManager().install()) # selenium이 지원하는 chrome 버전 웹드라이버 파일 넣어주기
```

```python
# 무한 스크롤 https://velog.io/@insutance/%EB%AC%B4%ED%95%9C-%EC%8A%A4%ED%81%AC%EB%A1%A4-%ED%8E%98%EC%9D%B4%EC%A7%80-%ED%81%AC%EB%A1%A4%EB%A7%81
import time
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException


def scrolling():
    elem = driver.find_element(By.TAG_NAME, "body")
    # 현재 body 부분의 높이를 last_height
    last_height = driver.execute_script("return document.body.scrollHeight")
    
    # 반복문을 통해 최하단까지 내려간다.
    while True:
        scroll_down = 0
        while scroll_down < 10:
            elem.send_keys(Keys.PAGE_DOWN)  # page down 버튼을 누르게해서 계속 페이지 하단으로 내려감
            time.sleep(1)
            scroll_down += 1
        
        # page down 10번 후 새로운 높이를 new_height에 담아줌
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:  # 높이가 변하지 않았다면 break
            break
        
        # 로딩이 완료될 때까지 대기
        try:
            WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.TAG_NAME, "body")))
        except TimeoutException:
            pass
        
        last_height = new_height
    return
```

```python

# platfoms = ['netflix','disney-plus','wavve','watch','apple-tv-plus']
genres = ['act','ani','cmy','crm','doc','drm','fnt','hst','hrr','fml','msc','trl','rma','scf','spt','war','wsn','rly','atp']

links = dict()
for genre in genres:
    driver.get(url='https://www.justwatch.com/kr/%EB%8F%99%EC%98%81%EC%83%81%EC%84%9C%EB%B9%84%EC%8A%A4/' + 'netflix' + '?=' + genre)   # netflix 자리에 다른 ott 넣어주기
    WebDriverWait(driver, 3).until(EC.presence_of_element_located((By.TAG_NAME, "body")))
    scrolling()
    num = 1
    while True:
        try:
            elem = driver.find_element(By.XPATH,f'//*[@id="base"]/div[3]/div/div[2]/div[1]/div/div[{num}]')
            elem_link = elem.find_element(By.CLASS_NAME,'title-list-grid__item--link')
            link = elem_link.get_attribute('href')
            title = link.split('/')[-1]
            print(link)
            links[title] = link
            num += 1
        except:
            break
```
