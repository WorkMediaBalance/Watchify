```python
import pandas as pd
import numpy as np
import os

df_ntf = pd.read_csv('./dataset/netflix_fin.csv', encoding='utf-8')
df_wtc = pd.read_csv('./dataset/watcha_fin.csv', encoding='utf-8')
df_wvv = pd.read_csv('./dataset/wavve_fin.csv', encoding='utf-8')
df_dnp = pd.read_csv('./dataset/disney_plus_fin.csv', encoding='utf-8')
```

```python
df = pd.concat([df_ntf, df_wtc, df_wvv, df_dnp], ignore_index=True)
del df['Unnamed: 0']
df
```

```python
from collections import defaultdict
dictionary = defaultdict(list)
for idx in df.index:
    title = df.loc[idx, 'title']
    ott = df.loc[idx, 'ott']
    dictionary[title].append(ott)

for key, value in dictionary.items():
    dictionary[key] = list(set(value))
print(dictionary)
print(len(dictionary))
```

```python
import os
savePath = os.getcwd() +'\\dataset'
title = list(dictionary.keys())
ott = list(dictionary.values())
df = pd.DataFrame({"title": title, "ott": ott})

# 본인이 맡은 파트의 ott 이름 + 시작 index + 끝 index 형식으로 파일 저장
df.to_csv(savePath+"\\tryingggg.csv", index=True, encoding='utf-8-sig')
```

```python
import pandas as pd
import re

# 숫자 부분 추출하는 함수 정의
def extract_numbers(string):
    # print('input : ',string)
    pattern = r'\d+\.*\d*'  # 정규식 패턴: 숫자와 옵션으로 소수점(.)이 포함된 패턴
    numbers = re.findall(pattern, ''.join(string))  # 정규식 패턴에 해당하는 부분 추출
    return [int(float(num)) for num in numbers]  # 추출된 문자열을 숫자형으로 변환하여 리스트 반환

# 'ott_y' 컬럼의 값을 숫자 리스트로 변환
df['ott_y'] = df['ott_y'].apply(extract_numbers)
```

```python
# NaN 값들을 전부 0으로 바꿔주기
df.fillna(0, inplace=True)
```

```python
df[['season', 'ott_x', 'episode_cnt']] = df[['season', 'ott_x', 'episode_cnt']].astype(int)
```

```python
import ast
import re

# 데이터프레임에서 'genre' 컬럼의 모든 값을 추출하여 리스트로 만듦
lst = df['genres'].tolist()

result = []
for elem in lst:
    # 문자열을 리스트로 변환
    elem.replace(' ','')
    lst_elem = ast.literal_eval(elem)
    # print(lst_elem)
    # 알파벳만 추출하여 리스트에 추가
    elem_result = []
    for sub_elem in lst_elem:
        elem_result.append(''.join(re.findall('[a-zA-Z]+', sub_elem)))
        elem_result.extend(re.findall('[ㄱ-ㅣ가-힣]+', sub_elem))
    res = []
    for e in elem_result:
        if e == '':
            continue
        else:
            res.append(e)
    # print(result)
    result.append(res)
len(result)
# 결과 리스트를 데이터프레임의 'genre' 컬럼으로 할당
df['genres'] = result
df
```
