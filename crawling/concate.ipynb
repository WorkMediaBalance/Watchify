```python
import pandas as pd
import numpy as np
import os

folders = os.listdir('./dataset/watcha')
print(folders)
```

```python
df_all = pd.DataFrame()
for i in range(len(folders)):
    if folders[i].split('.')[-1] == 'csv':
        file = './dataset/watcha/'+folders[i]
        df = pd.read_csv(file, encoding='utf-8')
        df_all = pd.concat([df_all, df])
```

```python
df_all
```

```python
savePath = os.getcwd() +'\\dataset'

# 본인이 맡은 파트의 ott 이름 + 시작 index + 끝 index 형식으로 파일 저장
df_all.to_csv(savePath+"\\watcha.csv", index=True, encoding='utf-8-sig')
```
