```python
import csv
import os
import pandas as pd
import numpy as np

df = pd.read_csv('./dataset/watcha.csv', encoding='utf-8')
length = len(df)
df["ott"] = np.ones((length,1))
df["ott"][:]= 2    # watcha
del df["Unnamed: 0"]
del df["Unnamed: 0.1"]
```

```python
savePath = os.getcwd() +'\\dataset'

# 본인이 맡은 파트의 ott 이름 + 시작 index + 끝 index 형식으로 파일 저장
df.to_csv(savePath+"\\watcha_fin.csv", index=True, encoding='utf-8-sig')
```
